```python
from keras.layers import LSTM 
from keras.models import Sequential 
from keras.layers import Dense 

# 모델 선언 예시
model = Sequential() 
# 첫번째 LSTM 레이어
model.add(LSTM(30,return_sequences=True, input_shape=(data.shape[0], 1))) 
# 두번째 LSTM 레이어
model.add(LSTM(42,return_sequences=False))  
# 예측값 1개
model.add(Dense(1, activation='linear')) 
# 손실함수 지정 - 예측 값과 실제 값의 차이를 계산한다. MSE가 사용된다. 
# 최적화기 지정 - 일반적으로 adam을 사용한다.
model.compile(loss='mean_squared_error', optimizer='adam') 
model.summary()
```

```python
model.fit(data,target,epochs=1000)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```
