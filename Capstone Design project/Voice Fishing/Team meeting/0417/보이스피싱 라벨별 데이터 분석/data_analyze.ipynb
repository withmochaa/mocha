```python
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True)
```

```python
x_train = x_train.astype(np.float32)
x_test = x_test.astype(np.float32)
```

```python
x_train = tf.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
x_test = tf.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
print('변경된 차원 정보 :',x_train.shape)
```

```python
y_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense, Bidirectional, Conv1D, MaxPooling1D, Flatten
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import tensorflow as tf

model = tf.keras.Sequential()
# 입력층 - 1D CNN
model.add(Conv1D(filters=512, kernel_size=3, activation='relu', input_shape=(x_train.shape[1], x_train.shape[2])))
model.add(Conv1D(filters=256, kernel_size=3, activation='relu'))
model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.2))
# 추가 CNN 층
model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.2))
# 중간층 - LSTM
model.add(Bidirectional(LSTM(128, return_sequences=True)))
model.add(Bidirectional(LSTM(64, return_sequences=True)))
model.add(Bidirectional(LSTM(32, return_sequences=False)))
model.add(Dropout(0.2))
# 추가 Dense 층
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
# 출력층
model.add(Dense(1, activation='sigmoid'))

# 최적화 알고리즘과 손실 함수 설정
model.compile(optimizer=tf.keras.optimizers.Nadam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Early Stopping과 Model Checkpoint 적용
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True)

model.summary()
```

```python
y_train=y_train.astype("float")
y_test=y_test.astype("float")
```

```python
y_train
```

```python
training_record = model.fit(x_train, y_train,
                    epochs=50,
                    batch_size=4,
                    callbacks=[early_stopping, model_checkpoint],
                    validation_split=0.2,
                    shuffle=True)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.plot(training_record.history["loss"], label="Training Loss")
plt.plot(training_record.history["val_loss"], label="Validation Loss")
plt.legend()
fig.suptitle("Loss")
plt.show()
```
