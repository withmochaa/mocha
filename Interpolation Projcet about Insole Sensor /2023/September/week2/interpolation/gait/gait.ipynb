```python
xgb_data_loss_COPX=data_loss.iloc[:,0].copy()
xgb_data_loss_COPY=data_loss.iloc[:,1].copy()

lstm_data_loss_COPX=data_loss.iloc[:,0].copy()
lstm_data_loss_COPY=data_loss.iloc[:,1].copy()

dnn_data_loss_COPX=data_loss.iloc[:,0].copy()
dnn_data_loss_COPY=data_loss.iloc[:,1].copy()
```

```python
import numpy as np 

index=data_loss.index

loss_index_COPX=np.array(data_loss.index).reshape(-1,1)[data_loss.iloc[:,0].isna()]
loss_index_COPY=np.array(data_loss.index).reshape(-1,1)[data_loss.iloc[:,1].isna()]
```

```python
normal_index_COPX=np.setdiff1d(index,loss_index_COPX)
normal_index_COPY=np.setdiff1d(index,loss_index_COPY)
```

```python
print(loss_index_COPX.shape)
print(loss_index_COPY.shape)
```

```python
x_train_COPX=normal_index_COPX.reshape(-1,1)
y_train_COPX=data_loss.iloc[normal_index_COPX,0]
x_test_COPX=loss_index_COPX

x_train_COPY=normal_index_COPY.reshape(-1,1)
y_train_COPY=data_loss.iloc[normal_index_COPY,1]
x_test_COPY=loss_index_COPY
```

```python
from xgboost.sklearn import XGBRegressor

reg = XGBRegressor().fit(x_train_COPX,y_train_COPX)
```

```python
pred=reg.predict(x_test_COPX)
pred=pred.reshape(-1,1)

c=0
for i in loss_index_COPX:    
    xgb_data_loss_COPX[i]=pred[c]
    c+=1
```

```python
import matplotlib.pyplot as plt 

plt.plot(xgb_data_loss_COPX,label='imputation_xgb_COPX')
plt.plot(data_loss.iloc[:,0],label='original')
plt.legend()
```

```python
from xgboost.sklearn import XGBRegressor

reg = XGBRegressor().fit(x_train_COPY,y_train_COPY)
```

```python
pred=reg.predict(x_test_COPY)
pred=pred.reshape(-1,1)

c=0
for i in loss_index_COPY:    
    xgb_data_loss_COPY[i]=pred[c]
    c+=1
```

```python
plt.plot(xgb_data_loss_COPY,label='imputation_xgb_COPY')
plt.plot(data_loss.iloc[:,1],label='original')
plt.legend()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.layers import LSTM

model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(427, 1)))
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(x_train_COPX,y_train_COPY,epochs=100,batch_size=16)
pred=model.predict(x_test_COPX)
pred=pred.reshape(-1,1)
c=0
for i in loss_index_COPX:    
    lstm_data_loss_COPX[i]=pred[c]
    c+=1
```

```python
dnn_data_loss_COPX=data_loss.iloc[:,0].copy()
dnn_data_loss_COPY=data_loss.iloc[:,1].copy()
```

```python
from keras import models, layers
from tensorflow.keras import layers, models


def build_model():
    model = models.Sequential()
    model.add(layers.Dense(64, activation = 'relu', input_shape=(1, )))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model
```

```python
dnn=build_model()
dnn.fit(x_train_COPX,y_train_COPY,epochs=100,batch_size=1,verbose=0)

pred=dnn.predict(x_test_COPX)
pred=pred.reshape(-1,1)

c=0
for i in loss_index_COPX:    
    dnn_data_loss_COPX[i]=pred[c]
    c+=1
```

```python
plt.plot(dnn_data_loss_COPX,label='imputation_dnn_COPX')
plt.plot(data_loss.iloc[:,0],label='original')
plt.legend()
```

```python
dnn=build_model()
dnn.fit(x_train_COPY,y_train_COPY,epochs=100,batch_size=1,verbose=0)

pred=dnn.predict(x_test_COPY)
pred=pred.reshape(-1,1)

c=0
for i in loss_index_COPY:    
    dnn_data_loss_COPY[i]=pred[c]
    c+=1
```

```python
plt.plot(dnn_data_loss_COPY,label='imputation_dnn_COPY')
plt.plot(data_loss.iloc[:,1],label='original')
plt.legend()
```
