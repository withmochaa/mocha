```python
import librosa.display
import numpy as np  
import os
import cv2

sample_rate = 8000

def create_mel_spectrogram(file_list, dir_path):
    audio_data = []
    for file in file_list:
        x, sr = librosa.load(os.path.join(dir_path, file), sr=sample_rate)
        S = librosa.feature.melspectrogram(y=x, sr=sr, n_mels=128)
        log_S = librosa.power_to_db(S, ref=np.max)
        img = cv2.resize(log_S, (224, 224))
        img_rgb = np.repeat(img[..., np.newaxis], 3, axis=2)
        audio_data.append(img_rgb)
    return np.array(audio_data)
```
