#### 분류 score 계산

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.model_selection import KFold

def score(pred, y_test):
    print(' accuracy  : ', accuracy_score(y_test, pred))
    print(' f1-score  : ', f1_score(y_test, pred,average='weighted'))
    print(' recall    : ', recall_score(y_test, pred,average='weighted'))
    print(' precision : ', precision_score(y_test, pred,average='weighted'))
    return accuracy_score(y_test, pred), f1_score(y_test, pred,average='weighted'), recall_score(y_test, pred,average='weighted'), precision_score(y_test, pred,average='weighted')
```

```python
def result_append(score_result, scores):
    A=pd.DataFrame(scores, index=score_result.columns)
    print(A)
    for j in range(A.shape[0]):
        score_result.iloc[0,j]=A.iloc[j,0]
        
    
    return score_result


```

```python
score_result = pd.DataFrame(index=range(1),columns=['logistic_Acc', 'logistic_F1', 'logistic_Rec', 'logistic_Pre',
                                     'naivebayes_Acc', 'naivebayes_F1', 'naivebayes_Rec', 'naivebayes_Pre',
                                     'Randomforest_Acc', 'Randomforest_F1', 'Randomforest_Rec', 'Randomforest_Pre',
                                     'XGBClassifier_ACC', 'XGBClassifier_F1', 'XGBClassifier_REC','XGBClassifier_Pre',
                                     'SVM_ACC', 'SVM_F1', 'SVM_REC', 'SVM_REC']
                                     
                                     )
```

```python
score_result.columns
```

```python
def SVM(x_train, y_train, test_x, test_y):

    # 모델 분류 수행
    reg = svm.SVC(decision_function_shape='ovo')
    reg.fit(x_train, y_train)
    pred = reg.predict(test_x)
    pred

    # 분류 score 계산
    accuracy, f1_score, recall, precision = score(pred, test_y)
    return accuracy, f1_score, recall, precision
```

```python
def xgb_clf(x_train, y_train, test_x, test_y):

    # 모델 분류 수행
    reg = XGBClassifier(random_state=42)
    reg.fit(x_train, y_train)
    pred = reg.predict(test_x)
    pred

    # 분류 score 계산
    accuracy, f1_score, recall, precision = score(pred, test_y)
    return accuracy, f1_score, recall, precision
```

```python
def logistic_reg(x_train, y_train, test_x, test_y):

    # 모델 분류 수행
    reg = LogisticRegression(random_state=42, max_iter=500)
    reg.fit(x_train, y_train)
    pred = reg.predict(test_x)
    pred

    # 분류 score 계산
    accuracy, f1_score, recall, precision = score(pred, test_y)
    return accuracy, f1_score, recall, precision
```

```python
def naivebayes_clf(x_train, y_train, test_x, test_y):

    # 모델 분류 수행
    gnb = GaussianNB()
    gnb.fit(x_train, y_train)
    pred = gnb.predict(test_x)


    # 분류 score 계산
    accuracy, f1_score, recall, precision = score(pred, test_y)
    return accuracy, f1_score, recall, precision
```

```python
def randomforest_clf(x_train, y_train, test_x, test_y):

    # 모델 분류 수행
    clf = RandomForestClassifier(max_depth=16, random_state=42)
    clf.fit(x_train, y_train)
    pred = clf.predict(test_x)


    # 분류 score 계산
    accuracy, f1_score, recall, precision = score(pred, test_y)
    return accuracy, f1_score, recall, precision
```

```python
def total_model(x_train,y_train,test_x,test_y):

    
    
    score_list = []
    
    

    # Logistic Regression
    print('      ----------    Logistic Regression Result    ----------      ')
    logi1, logi2, logi3, logi4 = logistic_reg(x_train, y_train, test_x,test_y)
    score_list += [logi1, logi2, logi3, logi4]


    # Naive Bayes Classification
    print('      ---------- Naive Bayes Classification Result ----------      ')
    naive1, naive2, naive3, naive4 = naivebayes_clf(x_train, y_train, test_x,test_y)
    score_list += [naive1, naive2, naive3, naive4]


    # RandomForest Classification
    print('      ---------- RandomForest Classifcation Result ----------      ')
    rf1, rf2, rf3, rf4 = randomforest_clf(x_train, y_train, test_x,test_y)
    score_list += [rf1, rf2, rf3, rf4]
    
    
    # XGBoost Classifier
    print('      ----------    XGBoost Classifier Result    ----------      ')
    xgb1, xgb2, xgb3, xgb4 = xgb_clf(x_train, y_train, test_x,test_y)
    score_list += [xgb1, xgb2, xgb3, xgb4]
    
    # SVM Classifier
    print('      ----------    SVM Classifier Result    ----------      ')
    svm1, svm2, svm3, svm4 = SVM(x_train, y_train, test_x,test_y)
    score_list += [svm1, svm2, svm3, svm4 ]


    
    return score_list
```

```python
output = total_model(x_train,y_train,test_x,test_y)
score_result = result_append(score_result, output)
score_result=score_result.transpose()
```
