```python
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True)
```

```python
x_train
```

```python
x_train.shape
```

```python
x_train = x_train.astype(np.float32)
x_test = x_test.astype(np.float32)
```

```python
x_train = tf.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
x_test = tf.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
print('변경된 차원 정보 :',x_train.shape)
```

```python
y_test
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dense, Input, TimeDistributed, Flatten
from keras.layers.convolutional import Conv1D ,MaxPooling1D
from keras.layers import Bidirectional
from keras.layers import LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dropout, TimeDistributed, Dense, Flatten
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.regularizers import l2


model = Sequential()
model.add(Bidirectional(LSTM(30, return_sequences=True, kernel_regularizer=l2(0.001)), input_shape=(x_train.shape[1], x_train.shape[2])))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(20, return_sequences=True, kernel_regularizer=l2(0.001))))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(10, return_sequences=True, kernel_regularizer=l2(0.001))))
model.add(TimeDistributed(Dense(10, activation='relu')))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid', kernel_initializer='uniform'))


model.compile(optimizer='Nadam', loss='binary_crossentropy', metrics=['accuracy'])


early_stopping = EarlyStopping(monitor='val_loss', patience=10)
model_checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True)

model.summary()
```

```python
y_train=y_train.astype("float")
y_test=y_test.astype("float")
```

```python
y_train
```

```python
training_record = model.fit(x_train, y_train,
                    epochs=50,
                    batch_size=4,
                    callbacks=[early_stopping, model_checkpoint],
                    validation_split=0.2,
                    shuffle=True)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.plot(training_record.history["loss"], label="Training Loss")
plt.plot(training_record.history["val_loss"], label="Validation Loss")
plt.legend()
fig.suptitle("Loss")
plt.show()
```

```python
pred = model.predict(x_test)
```

```python
y_test
```

```python
predictions=np.array(pred)
predictions = (predictions > 0.5).astype(float)
predictions=predictions.reshape(-1,1)
```

```python
predictions
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle


# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, predictions)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(10, 5))

# ROC Curve 그리기
plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")
```

```python
# Precision-Recall Curve
from sklearn.metrics import average_precision_score
precision, recall, _ = precision_recall_curve(y_test, predictions)
average_precision = average_precision_score(y_test, predictions)

# PR Curve 그리기
plt.subplot(1, 2, 2)
plt.plot(recall, precision, color='blue', lw=2, label='Precision-Recall curve (area = %0.2f)' % average_precision)
plt.fill_between(recall, precision, step='post', alpha=0.2, color='blue')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('PRC')
plt.legend(loc="lower left")

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score
print('BILSTM 2 class Classfication의 정확도는: {0:.4f}'.format(accuracy_score(y_test, predictions)))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# 혼동 행렬 및 분류 보고서 출력
print('혼동 행렬:')
print(confusion_matrix(y_test, predictions))
```

```python
print('분류 보고서:')
report=classification_report(y_test, predictions)
print(report)
```
