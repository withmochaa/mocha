```python
data=pd.DataFrame(data_row['sentence'].str.split('.').tolist())
```

```python
data3=pd.DataFrame(data.fillna('0'))
```

```python
count=0
for i in range(data3.shape[0]):
    for j in range(data3.shape[1]):
        if data3.iloc[i,j]!='0':
            count=count+1
```

```python
y=pd.DataFrame(index=range(count),columns=['sentence'])
```

```python
y.shape
```

```python
num=0
for i in range(data3.shape[0]):
    for j in range(data3.shape[1]):
        if data3.iloc[i,j]!='0':
            y.iloc[num,0]=data3.iloc[i,j]
            num=num+1
```

```python
y
```

```python
data=pd.DataFrame(index=range(y.shape[0]),columns=['predict'])
```

```python
data=data.fillna('0')
```

```python
norm=pd.concat([y,data],axis=1)
```
