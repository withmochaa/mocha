```python
x_train = tf.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
test_x = tf.reshape(test_x,(test_x.shape[0],test_x.shape[1],1))
print('변경된 차원 정보 :',x_train.shape)

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense, Bidirectional, Conv1D, MaxPooling1D, Flatten
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import tensorflow as tf

model = tf.keras.Sequential()
# 입력층 - 1D CNN
model.add(Conv1D(filters=256, kernel_size=3, activation='relu', input_shape=(x_train.shape[1], x_train.shape[2])))
model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.5))
# 추가 CNN 층
model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
model.add(Conv1D(filters=16, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.5))
# 중간층 - LSTM
model.add(Bidirectional(LSTM(128, return_sequences=True)))
model.add(Bidirectional(LSTM(64, return_sequences=True)))
model.add(Bidirectional(LSTM(32, return_sequences=False)))
model.add(Dropout(0.5))
# 추가 Dense 층
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
# 출력층
model.add(Dense(3, activation='softmax'))

# 최적화 알고리즘과 손실 함수 설정
model.compile(optimizer=tf.keras.optimizers.Nadam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# Early Stopping과 Model Checkpoint 적용
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True)

model.summary()


```

```python
training_record = model.fit(x_train, y_train,
                    epochs=30,
                    batch_size=80,
                    callbacks=[early_stopping, model_checkpoint],
                    validation_split=0.2,
                    shuffle=True)
```

```python
model.save('CNN-BILSTM_model.h5')
```

```python
fig, ax = plt.subplots()
plt.plot(training_record.history["loss"], label="Training Loss")
plt.plot(training_record.history["val_loss"], label="Validation Loss")
plt.legend()
fig.suptitle("Loss")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle


n_classes = 3  # 클래스 개수

# 실제 레이블을 이진화
y_test_binarized = label_binarize(test_y, classes=[0, 1, 2])
y_score= model.predict(test_x)
```

```python
n_classes
```

```python
y_test_binarized
```

```python
# ROC 곡선 및 AUC 계산

fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
# PRC 계산
precision = dict()
recall = dict()
prc_auc = dict()
for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_test_binarized[:, i], y_score[:, i])
    prc_auc[i] = auc(recall[i], precision[i])
```

```python
# 클래스별 ROC 곡선 그리기
colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='ROC curve of class {0} (area = {1:0.2f})'.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic for multi-class')
plt.legend(loc="lower right")
plt.show()
```

```python
# 클래스별 PRC 곡선 그리기
for i, color in zip(range(n_classes), colors):
    plt.plot(recall[i], precision[i], color=color, lw=2,
             label='PRC curve of class {0} (area = {1:0.2f})'.format(i, prc_auc[i]))

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall curve for multi-class')
plt.legend(loc="lower left")
plt.show()
```
