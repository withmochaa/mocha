```python
process1_x_test=process1_test_row
process1_x_test=process1_x_test.drop(columns=['predict'])
process1_x_test=process1_x_test.drop(columns=['Unnamed: 0'])
process1_y_test=pd.DataFrame(process1_test_row['predict'])
```

```python
print(process1_x_train.shape)
print(process1_y_train.shape)
print(process1_x_test.shape)
print(process1_y_test.shape)
```

```python
import tensorflow as tf

x_train = tf.reshape(process1_x_train,(process1_x_train.shape[0],process1_x_train.shape[1],1))
test_x = tf.reshape(process1_x_test,(process1_x_test.shape[0],process1_x_test.shape[1],1))
```

```python
print(x_train.shape)
print(process1_y_train.shape)
print(test_x.shape)
print(process1_y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense, Bidirectional, Conv1D, MaxPooling1D, Flatten,ELU,Permute
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint,LearningRateScheduler
import tensorflow as tf
from tensorflow.keras.activations import elu


def lr_schedule(epoch, lr):
    if epoch > 5:
        return float(lr * tf.math.exp(-0.1))  
    return float(lr)  

fishing_detection_model = tf.keras.Sequential()

fishing_detection_model.add(Conv1D(filters=128, kernel_size=1, input_shape=(x_train.shape[1], x_train.shape[2])))
fishing_detection_model.add(ELU())

fishing_detection_model.add(Conv1D(filters=64, kernel_size=1))
fishing_detection_model.add(ELU())

fishing_detection_model.add(Conv1D(filters=32, kernel_size=1))
fishing_detection_model.add(ELU())

fishing_detection_model.add(MaxPooling1D(pool_size=1))

fishing_detection_model.add(Dropout(0.5))


fishing_detection_model.add(Permute((2, 1)))

fishing_detection_model.add(Bidirectional(LSTM(128, return_sequences=True)))
fishing_detection_model.add(Bidirectional(LSTM(64, return_sequences=True)))
fishing_detection_model.add(Bidirectional(LSTM(32, return_sequences=True)))
fishing_detection_model.add(Bidirectional(LSTM(16, return_sequences=False)))

fishing_detection_model.add(Dropout(0.5))

fishing_detection_model.add(Dense(128))
fishing_detection_model.add(ELU())

fishing_detection_model.add(Dropout(0.5))

fishing_detection_model.add(Dense(1, activation='sigmoid'))

fishing_detection_model.compile(optimizer=tf.keras.optimizers.AdamW(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_fishing_detection_model.keras', save_best_only=True)
lr_scheduler = LearningRateScheduler(lr_schedule)

fishing_detection_model.summary()
```

```python
training_record = fishing_detection_model.fit(x_train, process1_y_train,
                    epochs=10,
                    batch_size=32,
                    callbacks=[early_stopping, model_checkpoint,lr_scheduler],
                    validation_split=0.3,
                    shuffle=True)
```

```python
fishing_detection_model.save('fishing_detection_model(base v2).keras')
```

```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve, accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(8, 6))
plt.plot(training_record.history['loss'], label='train loss')
plt.plot(training_record.history['val_loss'], label='val loss')
plt.title('train and val loss')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
y_pred = fishing_detection_model.predict(test_x).ravel()
fpr, tpr, thresholds = roc_curve(process1_y_test, y_pred)
auc_value = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC = {auc_value:.2f}')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```

```python
precision, recall, _ = precision_recall_curve(process1_y_test, y_pred)
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label='precision-recall curve')
plt.title('PRC Curve')
plt.xlabel('Recall')
plt.ylabel('Precisio')
plt.legend(loc='lower left')
plt.show()
```

```python
y_pred_binary = np.where(y_pred > 0.5, 1, 0)  
accuracy = accuracy_score(process1_y_test, y_pred_binary)
print(f'accuracy score: {accuracy:.2f}')
```

```python
cm = confusion_matrix(process1_y_test, y_pred_binary)
fig, ax = plt.subplots(figsize=(6, 6))
cax = ax.matshow(cm, cmap=plt.cm.Blues)
fig.colorbar(cax)
plt.title('confusion matrix')
plt.xlabel('prediction value')
plt.ylabel('real value')
plt.xticks([0, 1], ['0', '1'])
plt.yticks([0, 1], ['0', '1'])
for (i, j), val in np.ndenumerate(cm):
    ax.text(j, i, f'{val}', ha='center', va='center')
plt.show()
```
