```python
import os
import pandas as pd
import random
import re

from tqdm.notebook import tqdm
from kss import split_sentences
```

```python
def get_상담_info():
    global total_text_list_800
    data_path = '데이터 저장경로'    # 데이터 저장 경로
    data_list = os.listdir(data_path)
    total_text_list  = []           # 개별 상담 내용
    total_text_count = []           # 개별 상담 문장 개수
    
    # 폴더에 접근해서 텍스트 수집
    for folder in tqdm(data_list):
        if folder == ".DS_Store":
            continue
        
        # 개별 상담 건수에 접근
        file_path = data_path + '/' + folder
        file_list = os.listdir(file_path)
        file_list.sort()
        file_list_py = [file for file in file_list if file.endswith('.txt')] 
        

        # 문장별로 나뉜 텍스트를 합침
        total_txt = ''
        for file_name in file_list_py:
            text_file = open(file_path+'/'+file_name, 'r', encoding='UTF8')
            line = text_file.readline()
            total_txt += line
            
        
    
        total_text_list.append(total_txt)
        
        total_text_count.append(len(file_list_py))
        
    # data_list, total_text_list, total_text_count 리스트에서 앞에서부터 800개 추출
    data_list_800 = data_list[:800]
    total_text_list_800 = total_text_list[:800]
    total_text_count_800 = total_text_count[:800]

    # 추출한 데이터로 DataFrame 생성
    counseling_전체_800 = pd.DataFrame({'인덱스': data_list_800, '내용': total_text_list_800, '문장 개수': total_text_count_800})

    return counseling_전체_800
```

```python
#문장분리
i = 0
for text in total_text_list_800:
    text = text.replace('n/', '')
    sentences = re.split(r'\.|\?', text)  # 가정: 문장은 마침표(.)로 구분되어 있다고 가정
    total_text_list_800[i] = sentences
    i += 1
```

```python
import re

def replace_expression(match):
    words = match.group(1).split('/')
    return words[0]

for i in range(len(first_data)):
    conversationList = first_data[i]
    for j in range(len(conversationList)):
        text = conversationList[j]
        text = text.replace('+', '')
        
        pattern = r'\(([^)]+)\)/\([^)]+\)'
        text = re.sub(pattern, replace_expression, text)
        conversationList[j] = text
    first_data[i] = conversationList
```
