```python
import pandas as pd


x_train=pd.read_csv('/content/drive/MyDrive/파이썬기반기계학습/COVID-19/x_train.csv')
x_test=pd.read_csv('/content/drive/MyDrive/파이썬기반기계학습/COVID-19/x_test.csv')
y_train=pd.read_csv('/content/drive/MyDrive/파이썬기반기계학습/COVID-19/y_train.csv')
submission=pd.read_csv('/content/drive/MyDrive/파이썬기반기계학습/COVID-19/sample_submission.csv')
```

```python
nul=[]

for col in x_test.columns:
  if x_train[col].isnull().sum()>326:
    x_train=x_train.drop(columns=col,axis=1)
    x_test=x_test.drop(columns=col,axis=1)
  elif x_test[col].isnull().sum()>38:
    x_train=x_train.drop(columns=col,axis=1)
    x_test=x_test.drop(columns=col,axis=1)
```

```python
x_train.info()
```

```python
x_test.info()
```

```python
train_mean=x_train.mean(axis=0)
test_mean=x_test.mean(axis=0)
```

```python
x_train=x_train.fillna(method='bfill')

x_test=x_test.fillna(method='ffill')
```

```python
x_train.isnull().sum()
```

```python
display(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

x_train=x_train.drop(labels='date',axis=1)
x_test=x_test.drop(labels='date',axis=1)


train_scaler=scaler.fit_transform(x_train)
test_scaler=scaler.fit_transform(x_test)

x_train=pd.DataFrame(data=train_scaler)
x_test=pd.DataFrame(data=test_scaler)
```

```python
y_train=y_train.drop(labels='date',axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor(n_estimators=5000)

model.fit(x_train,y_train)

pred=model.predict(x_test)
```
