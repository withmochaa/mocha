```python
def plot_slope(x,y,ax):
    xs=x[1:]-x[:-1]
    ys=y[1:]-y[:-1]
    ax.plot(x[1:],ys/xs,'b')
fig=plt.figure(figsize=(10,5),dpi=100)
ax=fig.add_subplot(1,1,1)
x=np.linspace(-3,3,100)
y=np.exp(-x**2)
ax.plot(x,y,'r')
plot_slope(x,y,ax)
```
