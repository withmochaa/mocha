```python
import pandas as pd

data=pd.read_excel('/Users/withmocha/Desktop/TEED/data/mocha/23.09.02/SQUAT.xlsx')
data_mean=pd.read_excel('/Users/withmocha/Desktop/TEED/data/mocha/23.09.02/SQUAT_mean.xlsx')
print(data_mean.shape)
print(data.shape)
```

```python
print(data.columns)
print(data_mean.columns)
```

```python
data_mean=data_mean.drop(columns=['Unnamed: 0'])
print(data_mean.shape)
```

```python
data=data.loc[:,['X COORDINATE_C(Left)','X COORDINATE_F(Center)','Y COORDINATE_C(Left)','Y COORDINATE_F(Center)']]
data_mean=data_mean.loc[:,['X COORDINATE_C(Left)','X COORDINATE_F(Center)','Y COORDINATE_C(Left)','Y COORDINATE_F(Center)']]
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
knnr_left_data_xleft=data.iloc[:,0].copy()
linear_left_data_xleft=data.iloc[:,0].copy()
re_f_left_data_xleft=data.iloc[:,0].copy()
svr_left_data_xleft=data.iloc[:,0].copy()
```

```python
knnr_left_data_xcenter=data.iloc[:,1].copy()
linear_left_data_xcenter=data.iloc[:,1].copy()
re_f_left_data_xcenter=data.iloc[:,1].copy()
svr_left_data_xcenter=data.iloc[:,1].copy()
```

```python
knnr_left_data_yleft=data.iloc[:,2].copy()
linear_left_data_yleft=data.iloc[:,2].copy()
re_f_left_data_yleft=data.iloc[:,2].copy()
svr_left_data_yleft=data.iloc[:,2].copy()
```

```python
knnr_left_data_ycenter=data.iloc[:,3].copy()
linear_left_data_ycenter=data.iloc[:,3].copy()
re_f_left_data_ycenter=data.iloc[:,3].copy()
svr_left_data_ycenter=data.iloc[:,3].copy()
```

```python
data
```

```python
import numpy as np 


loss_index_xleft=np.array(data.index).reshape(-1,1)[data.iloc[:,0].isna()]
loss_index_xcenter=np.array(data.index).reshape(-1,1)[data.iloc[:,1].isna()]
loss_index_yleft=np.array(data.index).reshape(-1,1)[data.iloc[:,2].isna()]
loss_index_ycenter=np.array(data.index).reshape(-1,1)[data.iloc[:,3].isna()]

```

```python
print(loss_index_xleft.shape)
print(loss_index_xcenter.shape)
print(loss_index_yleft.shape)
print(loss_index_ycenter.shape)
```

```python
index=data.index


normal_index_xleft=np.setdiff1d(index,loss_index_xleft)
normal_index_xcenter=np.setdiff1d(index,loss_index_xcenter)
normal_index_yleft=np.setdiff1d(index,loss_index_yleft)
normal_index_ycenter=np.setdiff1d(index,loss_index_ycenter)

```

```python
x_train_xleft=normal_index_xleft
y_train_xleft=data.iloc[normal_index_xleft,0]
x_test_xleft=loss_index_xleft


x_train_xcenter=normal_index_xcenter
y_train_xcenter=data.iloc[normal_index_xcenter,1]
x_test_xcenter=loss_index_xcenter


x_train_yleft=normal_index_yleft
y_train_yleft=data.iloc[normal_index_yleft,2]
x_test_yleft=loss_index_yleft


x_train_ycenter=normal_index_ycenter
y_train_ycenter=data.iloc[normal_index_ycenter,3]
x_test_ycenter=loss_index_ycenter
```

```python
knn=KNeighborsRegressor(n_neighbors=7)
lr=LinearRegression()
rfr=RandomForestRegressor()
svr=SVR(kernel='rbf')
```

```python
knn.fit(np.array(x_train_xleft).reshape(-1,1),np.array(y_train_xleft).reshape(-1,1))
pred=knn.predict(np.array(x_test_xleft).reshape(-1,1))
```

```python

c=0
for i in loss_index_xleft:    
    knnr_left_data_xleft.iloc[i]=pred[c]
    c+=1
```

```python
import matplotlib.pyplot as plt
plt.plot(knnr_left_data_xleft,label='imputation')
plt.plot(data_mean.iloc[:,0],label='original')
plt.legend()
```

```python
lr=LinearRegression()
lr.fit(np.array(x_train_xleft).reshape(-1,1),np.array(y_train_xleft).reshape(-1,1))
pred=lr.predict(np.array(x_test_xleft).reshape(-1,1))
```

```python
c=0
for i in loss_index_xleft:    
    linear_left_data_xleft.iloc[i]=pred[c]
    c+=1
```

```python
import matplotlib.pyplot as plt
plt.plot(linear_left_data_xleft,label='imputation')
plt.plot(data_mean.iloc[:,0],label='original')
plt.legend()
```

```python
rfr=RandomForestRegressor()
rfr.fit(np.array(x_train_xleft).reshape(-1,1),np.array(y_train_xleft).reshape(-1,1))
pred=rfr.predict(np.array(x_test_xleft).reshape(-1,1))
```

```python
c=0
for i in loss_index_xleft:    
    re_f_left_data_xleft.iloc[i]=pred[c]
    c+=1
```

```python
import matplotlib.pyplot as plt
plt.plot(re_f_left_data_xleft,label='imputation')
plt.plot(data_mean.iloc[:,0],label='original')
plt.legend()
```

```python
svr=SVR(kernel='rbf')
svr.fit(np.array(x_train_xleft).reshape(-1,1),np.array(y_train_xleft).reshape(-1,1))
pred=svr.predict(np.array(x_test_xleft).reshape(-1,1))
```

```python
c=0
for i in loss_index_xleft:    
    svr_left_data_xleft.iloc[i]=pred[c]
    c+=1
```

```python
import matplotlib.pyplot as plt
plt.plot(svr_left_data_xleft,label='imputation')
plt.plot(data_mean.iloc[:,0],label='original')
plt.legend()
```

```python
plt.plot(knnr_left_data_xleft,label='knn_imputation')
plt.plot(linear_left_data_xleft,label='linear_imputation')
plt.plot(re_f_left_data_xleft,label='rfr_imputation')
plt.plot(svr_left_data_xleft,label='svr_imputation')
plt.plot(data_mean.iloc[:,0],label='original_X COORDINATE_C(Left)')
plt.legend()
```
