```python
knnr_data_XCOORDINATE_D_heel=data.iloc[:,0].copy()
linear_data_XCOORDINATE_D_heel=data.iloc[:,0].copy()
rfr_data_XCOORDINATE_D_heel=data.iloc[:,0].copy()
svr_data_XCOORDINATE_D_heel=data.iloc[:,0].copy()
```

```python
knnr_data_YCOORDINATE_C_Left=data.iloc[:,1].copy()
linear_data_YCOORDINATE_C_Left=data.iloc[:,1].copy()
rfr_data_YCOORDINATE_C_Left=data.iloc[:,1].copy()
svr_data_YCOORDINATE_C_Left=data.iloc[:,1].copy()
```

```python
knnr_data_YCOORDINATE_D_Heel=data.iloc[:,2].copy()
linear_data_YCOORDINATE_D_Heel=data.iloc[:,2].copy()
rfr_data_YCOORDINATE_D_Heel=data.iloc[:,2].copy()
svr_data_YCOORDINATE_D_Heel=data.iloc[:,2].copy()
```

```python
knnr_data_COPX=data.iloc[:,3].copy()
linear_data_COPX=data.iloc[:,3].copy()
rfr_data_COPX=data.iloc[:,3].copy()
svr_data_COPX=data.iloc[:,3].copy()
```

```python
knnr_data_COPY=data.iloc[:,4].copy()
linear_data_COPY=data.iloc[:,4].copy()
rfr_data_COPY=data.iloc[:,4].copy()
svr_data_COPY=data.iloc[:,4].copy()
```

```python
import numpy as np 


loss_index_XCOORDINATE_D_heel=np.array(data.index).reshape(-1,1)[data.iloc[:,0].isna()]
loss_index_YCOORDINATE_C_Left=np.array(data.index).reshape(-1,1)[data.iloc[:,1].isna()]
loss_index_YCOORDINATE_D_Heel=np.array(data.index).reshape(-1,1)[data.iloc[:,2].isna()]
loss_index_COPX=np.array(data.index).reshape(-1,1)[data.iloc[:,3].isna()]
loss_index_COPY=np.array(data.index).reshape(-1,1)[data.iloc[:,4].isna()]
```

```python
print(loss_index_XCOORDINATE_D_heel.shape)
print(loss_index_YCOORDINATE_C_Left.shape)
print(loss_index_YCOORDINATE_D_Heel.shape)
print(loss_index_COPX.shape)
print(loss_index_COPY.shape)
```

```python
data.isnull().sum()
```

```python
index=data.index


normal_index_XCOORDINATE_D_heel=np.setdiff1d(index,loss_index_XCOORDINATE_D_heel)
normal_index_YCOORDINATE_C_Left=np.setdiff1d(index,loss_index_YCOORDINATE_C_Left)
normal_index_YCOORDINATE_D_Heel=np.setdiff1d(index,loss_index_YCOORDINATE_D_Heel)
normal_index_COPX=np.setdiff1d(index,loss_index_COPX)
normal_index_COPY=np.setdiff1d(index,loss_index_COPY)
```

```python
x_train_XCOORDINATE_D_heel=normal_index_XCOORDINATE_D_heel
y_train_XCOORDINATE_D_heel=data.iloc[normal_index_XCOORDINATE_D_heel,0]
x_test_XCOORDINATE_D_heel=loss_index_XCOORDINATE_D_heel


x_train_YCOORDINATE_C_Left=normal_index_YCOORDINATE_C_Left
y_train_YCOORDINATE_C_Left=data.iloc[normal_index_YCOORDINATE_C_Left,1]
x_test_YCOORDINATE_C_Left=loss_index_YCOORDINATE_C_Left


x_train_YCOORDINATE_D_Heel=normal_index_YCOORDINATE_D_Heel
y_train_YCOORDINATE_D_Heel=data.iloc[normal_index_YCOORDINATE_D_Heel,2]
x_test_YCOORDINATE_D_Heel=loss_index_YCOORDINATE_D_Heel


x_train_COPX=normal_index_COPX
y_train_COPX=data.iloc[normal_index_COPX,3]
x_test_COPX=loss_index_COPX

x_train_COPY=normal_index_COPY
y_train_COPY=data.iloc[normal_index_COPY,3]
x_test_COPY=loss_index_COPY
```

```python
knn=KNeighborsRegressor(n_neighbors=7)
lr=LinearRegression()
rfr=RandomForestRegressor()
svr=SVR(kernel='rbf')
```

```python
knn.fit(np.array(x_train_XCOORDINATE_D_heel).reshape(-1,1),np.array(y_train_XCOORDINATE_D_heel).reshape(-1,1))
pred=knn.predict(np.array(x_test_XCOORDINATE_D_heel).reshape(-1,1))

c=0
for i in loss_index_XCOORDINATE_D_heel:    
    knnr_data_XCOORDINATE_D_heel.iloc[i]=pred[c]
    c+=1
```

```python
lr.fit(np.array(x_train_XCOORDINATE_D_heel).reshape(-1,1),np.array(y_train_XCOORDINATE_D_heel).reshape(-1,1))
pred=lr.predict(np.array(x_test_XCOORDINATE_D_heel).reshape(-1,1))

c=0
for i in loss_index_XCOORDINATE_D_heel:    
    linear_data_XCOORDINATE_D_heel.iloc[i]=pred[c]
    c+=1
```

```python
rfr.fit(np.array(x_train_XCOORDINATE_D_heel).reshape(-1,1),np.array(y_train_XCOORDINATE_D_heel).reshape(-1,1))
pred=rfr.predict(np.array(x_test_XCOORDINATE_D_heel).reshape(-1,1))

c=0
for i in loss_index_XCOORDINATE_D_heel:    
    rfr_data_XCOORDINATE_D_heel.iloc[i]=pred[c]
    c+=1
```

```python
svr.fit(np.array(x_train_XCOORDINATE_D_heel).reshape(-1,1),np.array(y_train_XCOORDINATE_D_heel).reshape(-1,1))
pred=svr.predict(np.array(x_test_XCOORDINATE_D_heel).reshape(-1,1))

c=0
for i in loss_index_XCOORDINATE_D_heel:    
    svr_data_XCOORDINATE_D_heel.iloc[i]=pred[c]
    c+=1
```

```python
import matplotlib.pyplot as plt

plt.plot(knnr_data_XCOORDINATE_D_heel,label='knn_imputation')
plt.plot(linear_data_XCOORDINATE_D_heel,label='linear_imputation')
plt.plot(rfr_data_XCOORDINATE_D_heel,label='rfr_imputation')
plt.plot(svr_data_XCOORDINATE_D_heel,label='svr_imputation')
plt.plot(data_mean.iloc[:,0],label='original_XCOORDINATE_D_heel')
plt.legend()
```
